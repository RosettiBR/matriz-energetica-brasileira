```python
import pandas as pd

def combinar_tabelas(dataset):

    excel_file = pd.ExcelFile(dataset)

    nomes_abas = excel_file.sheet_names

    lista_dfs_ano = []

    for nome_aba in nomes_abas:
        print(f"Lendo abas: {nome_aba}")

        df = pd.read_excel(excel_file, sheet_name=nome_aba)


        df.dropna(how='all', subset=df.columns[1:], inplace=True)


        df['Ano'] = nome_aba

        lista_dfs_ano.append(df)

    df_final = pd.concat(lista_dfs_ano, ignore_index=True)

    return df_final


path_do_arquivo = '/content/drive/MyDrive/matriz-energetica-brasiliera/balanco_consolidados_2024_2005.xlsx'

df_combinado = combinar_tabelas(path_do_arquivo)


print(df_combinado.head())


```

```python
df_combinado = df_combinado.loc[:, ~df_combinado.columns.str.contains('^Unnamed')]
```

```python
display(df_combinado.head(50))
```

```python
df_combinado.columns = df_combinado.columns.str.strip()

for column in df_combinado.columns:
    if column not in ['CONTA', 'Ano']:
        df_combinado[column] = pd.to_numeric(df_combinado[column], errors='coerce')

print("Colunas convertidas para tipo numérico.")

df_combinado['Ano'] = pd.to_datetime(df_combinado['Ano'], format='%Y')
print("Coluna 'Ano' convertida para datetime.")

print(df_combinado.info())
```

```python
df_combinado.to_csv('/content/drive/MyDrive/matriz-energetica-brasiliera/balanco_consolidados_2024_2005_combinado.csv', index=False)
```
